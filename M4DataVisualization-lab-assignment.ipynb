# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')  # optional: for ggplot-like style


QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)

# plot histogram
df_ConvertedComp['ConvertedComp'].plot(kind='hist', color='r', figsize=(8, 5))
# add a title to the histogram
plt.title('Histogram of Converted Compensation of Respondents')
# add y-label
plt.ylabel('Number of Respondents')
# add x-label
plt.xlabel('ConvertedComp')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Respondent, Age
FROM master
"""
df_Age = pd.read_sql_query(QUERY,conn)

df_Age['Age'].plot(kind='box', figsize=(10, 7))

plt.title("Box plots of Age of Respondents")
plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
import seaborn as sns

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)

plot = sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)

plt.show()
```

```python
df.corr()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
df_bubble = pd.read_sql_query(QUERY,conn)

```

```python
#normalize Age data
norm_Age = (df_bubble['Age'] - df_bubble['Age'].min()) / (df_bubble['Age'].max() - df_bubble['Age'].min())
```

```python
df_bubble.plot(kind='scatter',
                x='WorkWeekHrs', 
                y='CodeRevHrs', 
                s=norm_Age*1000+10, #passes in weight
                alpha=0.5, #for transparency
                figsize=(14, 8),
                color='purple')

plt.title('Bubble Plot of Relationship between WorkWeekHrs and CodeRevHrs')
plt.xlabel('WorkWeekHrs'),
plt.ylabel('CodeRevHrs'),
plt.legend(['Age'], loc='upper right', fontsize='large')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as Respondent
from DatabaseDesireNextYear
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

df_pie = pd.read_sql_query(QUERY, conn)
df_pie.head()

```

```python
colors_list=['black', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'pink']

df_pie['Respondent'].plot(kind='pie',
            figsize=(6,8), 
            autopct='%1.1f%%', 
            labels=None, 
            startangle=90, 
            colors=colors_list, 
            shadow=True, 
            pctdistance=1.12)

plt.legend(labels=df_pie['DatabaseDesireNextYear'], loc='upper right')
plt.title('Top 5 Databases Respondents Wish To Learn Next Year')
plt.axis('equal')

plt.show()
```

```python
#Create a pie chart of top Five languages respondents desire to learn next year

QUERY = """
SELECT LanguageDesireNextYear, COUNT(*) as Respondent 
FROM LanguageDesireNextYear
group by LanguageDesireNextYear
order by count(LanguageDesireNextYear) DESC LIMIT 5
"""
df_pie = pd.read_sql_query(QUERY,conn)

df_pie.head()

```

```python
colors_list=['black', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'pink']

df_pie['Respondent'].plot(kind='pie',
            figsize=(6,8), 
            autopct='%1.1f%%', 
            labels=None, 
            startangle=90, 
            colors=colors_list, 
            shadow=True, 
            pctdistance=1.12)

plt.legend(labels=df_pie['LanguageDesireNextYear'], loc='upper right')
plt.title('Top 5 Languages Respondents Wish To Learn Next Year')
plt.axis('equal')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
order by Age DESC
"""
df_stacked = pd.read_sql_query(QUERY, conn)
df_stacked

```

```python
df_stacked.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind='bar', stacked=True, figsize=(20, 10))
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35',
          loc='center',pad=20, fontsize=20, color='green')
plt.xlabel('CodeRevHrs', fontsize = 16)
plt.ylabel('WorkWeekHrs', fontsize = 16)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, ncol=1, title='Age')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
order by Age DESC
"""
df_line = pd.read_sql_query(QUERY,conn)
df_line
```

```python
df_line = df_line.groupby('Age').median()

df_line.plot(kind='line', color='green', figsize=(12, 6))

plt.title('Median ConvertedComp for all Ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

QUERY = """
SELECT MainBranch, COUNT(*) as Respondent
from master
group by MainBranch

"""
df_bar = pd.read_sql_query(QUERY,conn)
df_bar.head()

```

```python
df_bar.sort_values(["Respondent"], ascending=True, inplace=True)
# generate plot
df_bar = df_bar.set_index('MainBranch')
df_bar.plot(kind='barh', figsize=(20, 5), color='green')
plt.ylabel('MainBranch')
plt.title('Total Respondent Based on Main Branch')

plt.show()
```

```python
#Create a horizontal bar chart using column DevType
QUERY = """
SELECT *
FROM DevType
"""
df_bar = pd.read_sql_query(QUERY,conn)
df_bar.head()
```

```python
df_bar = df_bar.groupby('DevType', axis=0).count()

df_bar.sort_values(["Respondent"], ascending=True, inplace=True)
# generate plot
df_bar.plot(kind='barh', figsize=(20, 5), color='green')
plt.xlabel('Total')
plt.title('Total Respondent Based on Developer Type')

plt.show()
```
